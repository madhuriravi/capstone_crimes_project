# Req.No:1..Population of each state from 2001 to 2012

```python
import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population")
df
```

```python
import requests
from bs4 import BeautifulSoup
res=requests.get("https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population")
soup=BeautifulSoup(res.text,"html.parser")

data=soup.find_all("table")[1]
df=pd.read_html(str(data))
df[0]
```

```python
df = df[0]
df.to_csv('my_data.csv')
```

```python
df1=pd.read_csv("my_data.csv")
df1
```

```python
df2=df1[["List of states and union territories of India by population(2011 Census of India).1","List of states and union territories of India by population(2011 Census of India).2"]]
df2
```

df2 is population of each state  from year 2001 to 2012  and index 37 is Total population of india

# Req No:2..Literacy Rate in each state

```python
import requests
from bs4 import BeautifulSoup
res=requests.get("https://thingsinindia.in/literacy-rate-in-india/")
soup=BeautifulSoup(res.text,"html.parser")

data1=soup.find_all("table")
df3=pd.read_html(str(data1))
df3[0]
```

```python
df4=df3[0]
df4.to_csv('my_data1.csv')
```

```python
df4=pd.read_csv("my_data1.csv")
df4
```

```python
df4.rename(columns={"0":"State","1":"Literacy Rate"},inplace=True)
df4
```

```python
df4.drop([0,37],inplace=True)
df4
```

df4 is the Literacy rate of each state in India from year 2001 to 2012
